# **Contador de Monedas Automatico**
For this part, an automatic coin counting system was developed using the HoughCircles function for circles, which was used to determine the size of the coins and according to the size, the value of the coin was assigned and after detecting a certain amount of coins of the same value proceed to add the amount of bolivianos of the number of coins detected before using the coin count a bilateral filter is placed to soften the edges and eliminate noise from the image, a Gaussian filter could be used as a better implementation but in this case we wanted to use a bilateral filter to vary the content


```python
import numpy as np
import matplotlib.pyplot as plt
from google.colab import files as FILE
import os, requests
import cv2 as cv
from matplotlib import pyplot as plt
import matplotlib.image as mpimg
import numpy as np
import cv2 
%matplotlib inline

img = cv2.imread('menoy2.jpeg')
one = cv2.cvtColor(img, cv2.COLOR_BGR2RGB)
gray = cv2.cvtColor(img, cv2.COLOR_BGR2GRAY)
blurred =cv2.bilateralFilter(gray,10,50,50)
circles = cv2.HoughCircles(blurred, cv2.HOUGH_GRADIENT, 1, 50, param1=10, param2=50, minRadius=50, maxRadius=65)
if circles is not None:
    circles = np.uint16(np.around(circles))
    for i in circles[0,:]:
        cv2.circle(img, (i[0], i[1]), i[2], (0, 255, 0), 2)
total = list(circles.shape)[1]
ta=total/10
plt.imshow(one)
```
